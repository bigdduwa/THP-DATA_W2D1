# Récupération de l'adresse mail de tous les députés
Set-up : j'importe les librairies nécessaires pour le projet (bs4 = Beautiful Soup, permettant le scraping) et je fais la première requête avec requests.get(url) qui me permettra de chercher les éléments convoités dans le HTML

```python
import bs4
import requests
import pandas as pd

page_accueil = requests.get("https://www2.assemblee-nationale.fr/deputes/liste/tableau")
page_accueil
```

Je crée une liste appelée all_links regroupant les liens de toutes les pages des fiches personnelles des députés à partir de la page les recensant tous.

```python
soup = bs4.BeautifulSoup(page_accueil.text, 'html.parser')

all_links = [tag['href'] for tag in soup.select('td a[href]')]
len(all_links)
```

Maintenant j'utilise cette liste avec les bouts d'url pour créer une nouvelle liste contenant à chaque fois l'url complète (boucle avec début d'url + fin d'url)

```python
token = 'https://www2.assemblee-nationale.fr'
def get_pages(token, id):
    pages = []
    for i in range(1,id):
        j = token + all_links[i]
        pages.append(j)
    return pages
pages = get_pages(token,len(all_links))

pages
```

```python
def email_extractor(url):
  all_emails = []
  for i in pages:
    response = requests.get(i) #J'ouvre chaque lien un à un
    soup2 = bs4.BeautifulSoup(response.text, 'html.parser')
    new_email = [tag['href'] for tag in soup2.select('li a[href^=mailto]')] #Je cherche l'adresse email
    new_email = new_email[0] #Je sélectionne seulement le premier élément parce que la ligne de code au-dessus trouve deux adresses emails
    new_email = new_email[7:100] #Je supprime le 'mailto:' présent au début de chaque url
    all_emails.append(new_email) #J'ajoute chaque email à la liste
  return all_emails

print(email_extractor(pages))

```
