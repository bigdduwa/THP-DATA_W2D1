```python
import sqlite3
import pandas as pd

banque_db = sqlite3.connect("C:/Users/Victoire/OneDrive/Documents/THP/THP_DATA/Jour_6/data/py066-ma_banque.db")

print("1. Le nom et le prénom de tous les clients :")
nom_prenom = banque_db.execute("SELECT nom, prenom from client")
for row in nom_prenom:
  print(row[0], row[1])

banque_db.close
```

```python
print("Le nom et le prénom des clients habitant à Paris.")

nom_prenom_paris = banque_db.execute("SELECT nom, prenom, ville from client WHERE ville IS 'Paris'")
for row in nom_prenom_paris:
  print(row[2], ":",row[0], row[1])

banque_db.close
```

```python
print("Les identifiants des comptes de type Livret A :")
livret_a = banque_db.execute("SELECT idcompte FROM compte WHERE type is 'Livret A' ")
for row in livret_a:
  print(row[0])
```

```python
print("Les identifiants des opérations de débit sur le compte d’identifiant égal à 1 :")
op_debit_1 = banque_db.execute("SELECT idop FROM operation WHERE idcompte IS '1' AND montant LIKE '-%'")
for row in op_debit_1:
  print(row[0])
```

```python
print("Sans doublon, les identifiants des propriétaires de livret A, classés par ordre croissant :")

id_op_livreta = banque_db.execute("SELECT DISTINCT idproprietaire FROM compte WHERE type IS 'Livret A' ORDER BY idproprietaire ASC")

for row in id_op_livreta:
  print(row[0])
```

```python
print("L’identifiant des clients n’ayant pas de livret A :")

id_op_no_livreta = banque_db.execute("SELECT DISTINCT idproprietaire FROM compte WHERE type IS NOT 'Livret A' GROUP BY type ORDER BY idproprietaire ASC")
for row in id_op_no_livreta:
  print(row[0])

#?????????????????????????????????????
```

```python
print("L’identifiant de compte et le type de compte des clients habitant à Paris :")

id_compte_type_paris = banque_db.execute("SELECT idclient,type FROM client,compte WHERE client.idclient = compte.idproprietaire AND ville IS 'Paris' ")

for row in id_compte_type_paris:
  print(row)
```

```python
print("Liste des comptes et les types de compte de Dumbledore :")

dumbledore = banque_db.execute("SELECT type FROM client,compte WHERE client.idclient = compte.idproprietaire AND nom IS 'Dumbledore' ")

for row in dumbledore:
  print(row)
```

```python
print("Nombre de clients par ville, classé par ordre alphabétique de villes :")

city_count = banque_db.execute("SELECT ville,count(nom) FROM client GROUP BY ville ORDER BY ville ASC")

for row in city_count:
  print(row)
```

```python
print("Donner la ville ayant le plus de clients.")

city_max = banque_db.execute("SELECT ville,count(nom) FROM client GROUP BY ville ORDER BY count(nom) DESC LIMIT 1")

for row in city_max:
  print(row)
```

```python
print("Trouver le nombre d’opérations effectuées sur chaque compte.")
op_count_by_account = banque_db.execute("SELECT idcompte, count(idop) FROM operation GROUP BY idcompte ORDER BY idcompte ASC")

for row in op_count_by_account:
  print("compte n°",row[0],"nombre d'opérations :",row[1])
```

```python
print("Trouver le nombre maximum d’opérations effectuées sur un compte.")

op_count_by_account_max = banque_db.execute("SELECT idcompte, count(idop) FROM operation GROUP BY idcompte ORDER BY count(idop) DESC LIMIT 1")

for row in op_count_by_account_max:
  print("nombre d'opérations max :",row[1])
```
